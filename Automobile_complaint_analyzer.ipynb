```python
import pandas as pd
import numpy as np
import nltk
import pickle
import string
```

```python
from nltk.corpus import stopwords
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , classification_report
```

```python
nltk.download('stopwords')
```

```python
df = pd.read_excel('/content/automobile_complaint_dataset.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
x=df['Complaint_Text']
```

```python
y_system = df['Technical_System']
```

```python
y_severity = df['Severity_Level']
```

```python
stop_words = set (stopwords.words('english'))
```

```python
def clean_text(text) :
  text=text.lower()
  text = ''.join(char for char in text if char not in string.punctuation)
  text=text.split()
  text = [word for word in text if word not in stop_words]
  return ' '.join(text)
```

```python
x_clean = x.apply(clean_text)
```

```python
tfidf_vectorizer = TfidfVectorizer()
x_tfidf = tfidf_vectorizer.fit_transform(x_clean)
```

```python
X_train_system, X_test_system, y_train_system, y_test_system = train_test_split(x_tfidf, y_system, test_size=0.2, random_state=42)
```

```python
X_train_severity, X_test_severity, y_train_severity, y_test_severity = train_test_split(x_tfidf, y_severity, test_size=0.2, random_state=42)
```

```python
x_clean.head()
```

```python
vectorizer = TfidfVectorizer()
x_tfidf = vectorizer.fit_transform(x_clean)
```

```python
X_train, X_test, y_train_system, y_test_system = train_test_split(
    x_tfidf,
    y_system,
    test_size=0.2,
    random_state=42
)

_, _, y_train_severity, y_test_severity = train_test_split(
    x_tfidf,
    y_severity,
    test_size=0.2,
    random_state=42
)
```

```python
model_system = LogisticRegression(max_iter=1000)
model_system.fit(X_train, y_train_system)
```

```python
model_severity = LogisticRegression(max_iter=1000)
model_severity.fit(X_train, y_train_severity)
```

```python
y_pred_system = model_system.predict(X_test)
```

```python
print("Technical System Model Performance:")
print(f"Accuracy: {accuracy_score(y_test_system, y_pred_system):.4f}")
print("Classification Report:")
print(classification_report(y_test_system, y_pred_system))
```

```python
y_pred_severity = model_severity.predict(X_test)
```

```python
print("Severity Level Model Performance:")
print(f"Accuracy: {accuracy_score(y_test_severity, y_pred_severity):.4f}")
print("Classification Report:")
print(classification_report(y_test_severity, y_pred_severity))
```

```python
system_model = LogisticRegression(max_iter=1000)
system_model.fit(X_train,y_train_system)
```

```python
system_predictions = system_model.predict(X_test)
severity_predictions = model_severity.predict(X_test)
```

```python
print("SYSTEM MODEL ACCURACY")
print(accuracy_score(y_test_system, system_predictions))

print("\nSYSTEM CLASSIFICATION REPORT")
print(classification_report(y_test_system, system_predictions))
```

```python
print("SEVERITY MODEL ACCURACY")
print(accuracy_score(y_test_severity, severity_predictions))

print("\nSEVERITY CLASSIFICATION REPORT")
print(classification_report(y_test_severity, severity_predictions))
```

```python
from flask import Flask, request, jsonify

import pickle

import string

from nltk.corpus import stopwords
from threading import Thread
```

```python
system_model = pickle.load(open('system_model.pkl', 'rb'))

severity_model = pickle.load(open('severity_model.pkl', 'rb'))

vectorizer = pickle.load(open('vectorizer.pkl', 'rb'))
```

```python
app = Flask(__name__)
```

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):

    text = text.lower()

    text = ''.join(char for char in text if char not in string.punctuation)

    words = text.split()

    words = [word for word in words if word not in stop_words]

    return ' '.join(words)
```

```python
@app.route('/predict', methods=['POST'])

def predict():

    data = request.get_json()

    complaint = data['complaint']

    cleaned = clean_text(complaint)

    vector = vectorizer.transform([cleaned])

    probabilities = system_model.predict_proba(vector)

    confidence = max(probabilities[0]) * 100

    system_prediction = system_model.predict(vector)[0]

    severity_prediction = severity_model.predict(vector)[0]

    response = {
        'technical_system': system_prediction,
        'severity': severity_prediction,
        'confidence_score': round(confidence, 2)
    }

    return jsonify(response)
```
